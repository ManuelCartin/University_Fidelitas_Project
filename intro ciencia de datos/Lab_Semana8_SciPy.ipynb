```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import shapiro
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = 'sample_data/netflix_titles_nov_2019.csv'
data = pd.read_csv(file_path)

# Mostrar las primeras filas del dataframe
print(data.head())

```

```python

# Extraer la columna 'release_year' y convertirla a un array de NumPy
release_years = data['release_year'].values
release_years_array = np.array(release_years)

# Calcular estadísticas básicas
mean_year = np.mean(release_years_array)
median_year = np.median(release_years_array)
mode_year = stats.mode(release_years_array)[0]

print("Mean year:", mean_year)
print("Median year:", median_year)
print("Mode year:", mode_year)

```

```python

# Realizar la prueba de Shapiro-Wilk para la normalidad
stat, p = shapiro(release_years_array)
print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpretar el resultado
alpha = 0.05
if p > alpha:
    print('No se rechaza la hipótesis nula (H0)')
else:
    print('Se rechaza la hipótesis nula (H0)')
```

```python

# Filtrar los datos por tipo
movies = data[data['type'] == 'Movie']['release_year'].values
tv_shows = data[data['type'] == 'TV Show']['release_year'].values

# Realizar el t-test
t_stat, p_value = stats.ttest_ind(movies, tv_shows)

print('t-statistic:', t_stat)
print('p-value:', p_value)

# Interpretar el resultado
if p_value < alpha:
    print('Hay una diferencia significativa entre las medias')
else:
    print('No hay suficiente evidencia para decir que hay una diferencia significativa entre las medias')
```
