```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Leer el archivo CSV
data = pd.read_csv('sample_data/netflix_titles_nov_2019.csv')

print(data.head())
```

```python
# Extraer la columna 'release_year' y convertirla a un array de NumPy
release_years = data['release_year'].values
release_years_array = np.array(release_years)
print(release_years_array)
```

```python
# Calcular estadísticas básicas
mean_year = np.mean(release_years_array)
median_year = np.median(release_years_array)
min_year = np.min(release_years_array)
max_year = np.max(release_years_array)

print("Mean year:", mean_year)
print("Median year:", median_year)
print("Min year:", min_year)
print("Max year:", max_year)
```

```python
# Encontrar años únicos y su frecuencia
unique_years, counts = np.unique(release_years_array, return_counts=True)
print("Unique years:", unique_years)
print("Counts:", counts)
```

```python
# Crear un histograma de los años de lanzamiento
plt.hist(release_years_array, bins=np.arange(min_year, max_year+1)-0.5, edgecolor='black')
plt.title('Histogram of Release Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```
