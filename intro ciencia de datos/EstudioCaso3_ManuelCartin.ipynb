1.	Importe las librerías pandas, matplotlib y seaborn.
2.	Carga y revisión inicial de los datos para garantizar su calidad.
    -	Carga el archivo CSV en un DataFrame
    -	Muestra los primeros registros del DataFrame


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Leer el archivo CSV
df = pd.read_csv('sample_data/ventas_ejemplo.csv')

# Mostrar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())
```

3.	Cálculo de medidas estadísticas clave (media y mediana) para las ventas.
  -	Calcula la media de la columna ventas y muéstralo en pantalla
  -	Calcula la mediana de la columna ventas y muéstralo en pantalla


```python
# Calcular la media de la columna 'ventas'
media_ventas = df['Ventas'].mean()
print("\nMedia de ventas:")
print(media_ventas)

# Calcular la mediana de la columna 'ventas'
mediana_ventas = df['Ventas'].median()
print("\nMedia de ventas:")
print(mediana_ventas)
```

4.	Agrupación y análisis de ventas totales por región y por mes.
  	- Imprime en pantalla el total de las ventas totales por región
	  -  Imprime en pantalla el total de las ventas por mes


```python
# Agrupar por región y sumar las ventas
ventas_por_region = df.groupby("Region")["Ventas"].sum()
print("\nTotal de ventas por región:")
print(ventas_por_region)
```

```python
# Agrupar por región y sumar las ventas
ventas_por_mes = df.groupby("Mes")["Ventas"].sum()
print("\nTotal de ventas por Mes:")
print(ventas_por_mes)
```

5.	Creación de gráficos para presentar los hallazgos:
-	Gráfico de barras de ventas por región.
-	Gráfico de tendencia de ventas por mes.
-	Boxplot de distribución de ventas.
-	Gráfico de dispersión de ventas por mes y región.


```python
# Crear el gráfico de barras ventas por region
plt.figure(figsize=(8, 5))
ventas_por_region.plot(kind='bar', color=['blue', 'red', 'green', 'purple'])
plt.title("Total de Ventas por Región")
plt.xlabel("Región")
plt.ylabel("Total de Ventas")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(10, 5))
ventas_por_mes.plot(kind='line', marker='o', color='blue', linestyle='-')
plt.title("Total de Ventas por Mes")
plt.xlabel("Mes")
plt.ylabel("Total de Ventas")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Crear el gráfico de boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(y=df["Ventas"], color="lightblue")
plt.title("Distribución de Ventas")
plt.ylabel("Ventas")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Mes"], y=df["Ventas"], hue=df["Region"], palette="Set1", s=100)
plt.title("Distribución de Ventas por Mes y Región")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
