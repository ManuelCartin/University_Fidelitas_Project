```python
# Datos de ejemplo: pesos de 10 personas
pesos = np.array([70, 72, 68, 65, 74, 69, 71, 73, 67, 66])
```

```python
from scipy.stats import norm
import seaborn as sns
# Parámetros de la distribución normal estimada
media = np.mean(pesos)
desviacion = np.std(pesos)

# Generamos una distribución normal con la misma media y desviación estándar
x = np.linspace(media - 3*desviacion, media + 3*desviacion, 100)
y = norm.pdf(x, media, desviacion)

# Crear el gráfico
plt.figure(figsize=(8, 5))
sns.histplot(pesos, kde=True, bins=6, color='skyblue', stat="density", label="Datos observados")
#plt.plot(x, y, 'r-', label="Distribución Normal Estimada")
plt.axvline(media, color='black', linestyle='dashed', label=f"Media ({media:.2f})")

# Etiquetas y leyenda
plt.title("Distribución de los datos y curva normal estimada")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad")
plt.legend()
plt.grid()

# Mostrar gráfico
plt.show()
```

```python
from scipy.stats import shapiro

# Aplicamos la prueba de Shapiro-Wilk
stat, p = shapiro(pesos)

# Mostramos los resultados
print(f'Estadístico de Shapiro-Wilk: {stat:.4f}')
print(f'Valor p: {p:.4f}')

# Interpretación del resultado
if p > 0.05:
    print("Los datos parecen seguir una distribución normal.")
else:
    print("Los datos NO siguen una distribución normal.")
```

```python
# Realizar la prueba de Shapiro-Wilk para la normalidad
# Se esta aplicanso sobre los años de liberación, si existe una distribucióon normal.
stat, p = shapiro(release_years_array)
print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpretar el resultado
alpha = 0.05
if p > alpha:
    print('No se rechaza la hipótesis nula (H0)')
else:
    print('Se rechaza la hipótesis nula (H0)')
```

```python
# Filtrar los datos por tipo
peliculas = data[data['type'] == 'Movie']['release_year'].values
series = data[data['type'] == 'TV Show']['release_year'].values

# Realizar el t-test
t_stat, p_value = stats.ttest_ind(peliculas, series)

print('t-statistic:', t_stat)
print('p-value:', p_value)

# Interpretar el resultado
if p_value < alpha:
    print('Hay una diferencia significativa entre las medias')
else:
    print('No hay suficiente evidencia para decir que hay una diferencia significativa entre las medias')
```

```python

# Prueba de normalidad (Shapiro-Wilk)
stat_peliculas, p_peliculas = stats.shapiro(peliculas)
stat_series, p_series = stats.shapiro(series)

print(f"Prueba de normalidad - Películas: p = {p_peliculas}")
print(f"Prueba de normalidad - Series: p = {p_series}")

# Si ambos p > 0.05, podemos asumir normalidad
```

```python
# Prueba de igualdad de varianzas (Levene)
stat_var, p_var = stats.levene(peliculas, series)
print(f"Prueba de igualdad de varianzas: p = {p_var}")
```

```python
# Prueba t de Student
t_stat, p_val = stats.ttest_ind(peliculas, series, equal_var=(p_var > 0.05))
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p: {p_val:.4f}")
```

```python
# Interpretación
if p_val > 0.05:
    print("No hay suficiente evidencia para rechazar H0: Los años de lanzamiento son similares.")
else:
    print("Rechazamos H0: Hay una diferencia significativa en los años de lanzamiento.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Leer el archivo CSV
data = pd.read_csv('sample_data/starbucks-menu-nutrition-drinks.csv')

#
data.head()
```

```python
calories= data['Calories'].values
calories_array = np.array(calories)
carb= data['Carb. (g)'].values
carb_array = np.array(carb)
print(calories_array)
print(len(calories_array))
print(carb_array)
print(len(carb_array))
```

```python
# Calcular estadísticas básicas
mean_calo = np.mean(calories_array)
median_calo = np.median(calories_array)
min_calo = np.min(calories_array)
max_calo = np.max(calories_array)

print("Mean calories:", mean_calo)
print("Median calories:", median_calo)
print("Min calories:", min_calo)
print("Max calories:", max_calo)

# Calcular estadísticas básicas
mean_carb = np.mean(carb_array)
median_carb = np.median(carb_array)
min_carb = np.min(carb_array)
max_carb = np.max(carb_array)

print("Mean carb:", mean_carb)
print("Median carb:", median_carb)
print("Min carb:", min_carb)
print("Max carb:", max_carb)
```

```python
mode_calo = stats.mode(calories_array).mode

print("Mode calories:", mode_calo)

mode_carb = stats.mode(carb_array).mode

print("Mode carb:", mode_carb)
```
