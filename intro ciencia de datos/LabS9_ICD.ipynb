```python
import pandas as pd
import numpy as np

# Cargar el conjunto de datos
file_path = 'sample_data/vgsales.csv'
data = pd.read_csv(file_path)
print(data.head())
```

```python

# Seleccionar las columnas de interés
selected_columns = data[['Name', 'Platform', 'Year', 'Genre', 'Global_Sales']]
print(selected_columns.head())
```

```python

# Verificar valores faltantes
missing_values = selected_columns.isnull().sum()
print(missing_values)

```

```python
# Eliminar filas con valores faltantes en la columna 'Year'
cleaned_data = selected_columns.dropna(subset=['Year'])
print(cleaned_data)
```

```python
# Convertir la columna 'Year' a enteros
cleaned_data['Year'] = cleaned_data['Year'].astype(int)
print(cleaned_data)
```

```python
# Filtrar los datos para extraer juegos publicados en la plataforma 'Wii'
wii_games = cleaned_data[cleaned_data['Platform'] == 'Wii']
print(wii_games.head())

```

```python
# Filtrar los datos para extraer juegos publicados en el año '2002'
Year = data[data['Year'] > 2002]
print(Year.head())
```

```python

# Seleccionar las columnas de interés
selected_columns_country = data[[ 'NA_Sales' ,'EU_Sales', 'JP_Sales', 'Other_Sales']]
print(selected_columns_country.head())
```

```python
# Verificar valores faltantes
missing_values = selected_columns_country.isnull().sum()
print(missing_values)
```

```python
#eliminar filas con cvalores nulos
cleaned_data_country = selected_columns_country.dropna()
```

```python
#verificar valore nulos despues de eliminar
print("valores nulos despues de eliminar:")
print(cleaned_data_country.isnull().sum())
```

```python
ventas_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
promedios_ventas = Year[ventas_columns].mean()
print("promedios de evntas por region despues del 2002:")
print(promedios_ventas)
```
