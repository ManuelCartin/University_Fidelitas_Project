##Carga del data set
- Importa el archivo en un DataFrame de pandas y muestra las primeras 5 filas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import shapiro
# Leer el archivo
df = pd.read_csv('sample_data/vgsales.csv')
print(df.head())
```

#2. Selección estratégica de datos
-	Extrae solo las columnas Rank, Name, Platform, Genre, NA_Sales y Global_Sales, ya que serán clave para nuestro análisis.

```python
# Seleccionar las columnas de interés
selected_columns = df[['Rank', 'Name', 'Platform', 'Genre','NA_Sales', 'Global_Sales']]
print(selected_columns.head())

# Extraer la columna 'release_year' y convertirla a un array de NumPy
Global_Sales = df['Global_Sales'].values
Global_Sales_array = np.array(Global_Sales)
print(Global_Sales_array)
```

#Limpieza de datos
-	Elimina las filas con valores nulos para garantizar la calidad del análisis. (Si no la identificas puedes preguntarle al profesor)

```python
# Verificar valores faltantes
missing_values = selected_columns.isnull().sum()
print(missing_values)
```

```python
#eliminar valor nulo
# Eliminar filas con valores faltantes en la columna 'Year'
cleaned_data = selected_columns.dropna(subset=['NA_Sales'])
print(cleaned_data)
```

```python
# Verificar valores faltantes
missing_values_cleaned = cleaned_data.isnull().sum()
print(missing_values_cleaned)
```

#Transformacion de datos
-	Asegúrate de que la columna Global_Sales tenga el tipo de dato adecuado para facilitar cálculos numéricos. (float)

```python
# Convertir la columna 'Year' a enteros
cleaned_data['Global_Sales'] = cleaned_data['Global_Sales'].astype(float)
print(cleaned_data)
Global_Sales_float= Global_Sales_array.astype(float)
print(Global_Sales_float)
```

#Filtrado Inteligente
-	Selecciona todos los videojuegos lanzados exclusivamente en la plataforma PS3.


```python
# Filtrar los datos para extraer juegos publicados en la plataforma 'Wii'
PS3_Games = cleaned_data[cleaned_data['Platform'] == 'PS3']
print(PS3_Games.head())
```

#Calculo metricas Clave
-	Calcula el promedio de ventas globales para todos los videojuegos y analiza qué tan alto o bajo es este valor.

```python
# Calcular la media de la columna 'Global_Sales'
global_sales_mean = PS3_Games['Global_Sales'].mean()
print(f"La media de Global_Sales es: {global_sales_mean}")
```

#Hallazgo del más vendido
- Identifica el videojuego con mayores ventas globales y conviértelo en el rey de la lista.

```python
# Encuentra el índice del videojuego con más ventas globales
indice_rey = PS3_Games['Global_Sales'].idxmax()
videojuego_rey = PS3_Games.loc[indice_rey]
print("El videojuego con más ventas globales es:\n", videojuego_rey)

```

# Hallazgo del Menos Vendido:
-	Encuentra el videojuego con menos ventas globales y analiza posibles razones detrás de este resultado.


```python
# Encuentra el índice del videojuego con menos ventas globales
indice_menos_vendido = PS3_Games['Global_Sales'].idxmin()
videojuego_menos_vendido = PS3_Games.loc[indice_menos_vendido]
print("El videojuego con menos ventas globales es:\n", videojuego_menos_vendido)
```

# Análisis por Género:
- Agrupa los datos por género de videojuego y calcula la suma total de ventas para cada categoría.


```python
# Agrupar por 'Genre' y calcular la suma de ventas
ventas_por_genero = df.groupby('Genre')[['NA_Sales', 'Global_Sales']].sum()
print(ventas_por_genero)
```

# Búsqueda de un Clásico:
- Encuentra todos los videojuegos cuyo nombre contenga la palabra "Mario" y descubre qué tan influyente ha sido este personaje en la industria.


```python
# Filtrar los datos para extraer juegos publicados en la plataforma 'Wii'
Mario_Busqueda = PS3_Games[PS3_Games['Name'] == 'Mario']
print(Mario_Busqueda.head())
```
