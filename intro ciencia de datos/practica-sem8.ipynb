```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import shapiro
# Leer el archivo
df = pd.read_csv('sample_data/starbucks-menu-nutrition-drinks.csv')
print(df.head())
```

```python
# Extraer la columna 'release_year' y convertirla a un array de NumPy
calories = df['Calories'].values
calories_array = np.array(calories)
print(calories)
```

```python
# Extraer la columna 'release_year' y convertirla a un array de NumPy
carb = df['Carb. (g)'].values
carb_array = np.array(carb)
print(carb)
```

```python
# Calcular estadísticas básicas calories
mean_calories = np.mean(calories_array)
median_calories = np.median(calories_array)
min_calories = np.min(calories_array)
max_calories = np.max(calories_array)

print("Mean calories:", mean_calories)
print("Median calories:", median_calories)
print("Min calories:", min_calories)
print("Max calories:", max_calories)

# Calcular estadísticas carb
mean_carb = np.mean(carb_array)
median_carb = np.median(carb_array)
min_carb = np.min(carb_array)
max_carb = np.max(carb_array)

print("Mean carb:", mean_carb)
print("Median carb:", median_carb)
print("Min carb:", min_carb)
print("Max carb:", max_carb)
```

```python
# Realizar la prueba de Shapiro-Wilk para la normalidad calories
stat, p = shapiro(calories_array)
print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpretar el resultado
alpha = 0.05
if p > alpha:
    print('No se rechaza la hipótesis nula (H0)')
else:
    print('Se rechaza la hipótesis nula (H0)')

```

```python
# Realizar la prueba de Shapiro-Wilk para la normalidad carb
stat, p = shapiro(carb_array)
print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpretar el resultado
alpha = 0.05
if p > alpha:
    print('No se rechaza la hipótesis nula (H0)')
else:
    print('Se rechaza la hipótesis nula (H0)')
```
